```python
from PIL import Image, ImageFilter, ImageEnhance
import pytesseract
```

```python
# Load an image
img = Image.open('image.png')

# Extract text from image
text = pytesseract.image_to_string(img)

print(text)
```

```python
img = Image.open('Noisy_image.jpg')

img = img.convert('L')

img = img.filter(ImageFilter.MedianFilter())

enhancer = ImageEnhance.Contrast(img)
img = enhancer.enhance(2)

text = pytesseract.image_to_string(img)

print(text)
```

```python
img = Image.open('hq720.jpg')

img_gray = img.convert('L')

img_filtered = img_gray.filter(ImageFilter.MedianFilter(size=3))

enhancer = ImageEnhance.Contrast(img_filtered)
img_enhanced = enhancer.enhance(2.0)

custom_config = r'--oem 3 --psm 6'
text = pytesseract.image_to_string(img_enhanced, lang="eng")

print(text)
```
