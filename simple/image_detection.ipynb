```python
image = cv2.imread("room_1.jpg")
cv2.imshow("test IMage",image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
grayimg = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

cv2.imshow("egage Image",grayimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blurimg = cv2.GaussianBlur(grayimg,(1,1),0)
eadgeImage = cv2.Canny(blurimg,50,150)

cv2.imshow("blurimg Image",blurimg)
cv2.imshow("egage Image",eadgeImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
contours, hierarchy  = cv2.findContours(eadgeImage,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

output = image.copy()
cv2.drawContours(output,contours, -1, (0,255,0),2)

cv2.imshow("output Image",output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
```

```python

# === Step 1: Load the image ===
image_path = "room.jpg"  # change to your image path
img = cv2.imread(image_path)
if img is None:
    raise FileNotFoundError(f"Could not load image at {image_path}")
```

```python
# === Step 2: Convert to grayscale ===
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow("Gray Image",gray)
cv2.imshow("Normal Image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
# === Step 3: Denoise and enhance edges ===
blur = cv2.GaussianBlur(gray, (5, 5), 0)
edges = cv2.Canny(blur, 50, 150)

# cv2.imshow("blur Image",blur)
cv2.imshow("edges Image",edges)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
contours, hierarchy = cv2.findContours(edges,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
output = img.copy()
cv2.drawContours(output, contours, -1, (0, 255, 0), 2)  # Green contours

cv2.imshow("Contours", output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
